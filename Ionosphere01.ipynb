```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("ionosphere.data")
df1
```

```python
# adding labels to dataset

label_column = []
for i in range(1,36):
  label_column.append("col "+str(i))

label_column
```

```python
# df is dataframe
df = pd.read_csv("ionosphere.data", names = label_column)
df
```

```python
df['col 35'].unique()
```

```python
df["col 35"] = (df["col 35"] == "g").astype(int)
```

```python
df
```

```python
for label in label_column[:-1]:
  plt.hist(df[df["col 35"]==1][label], color = "green", label = "Good", alpha = 0.5, density = True)
  plt.hist(df[df["col 35"]==0][label], color = "red", label = "Bad", alpha = 0.5, density = True)
  plt.title(label)
  plt.legend()
  plt.show()
```
